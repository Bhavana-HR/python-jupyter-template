```python
# Importing necessary libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
# Calculating y values as the square of x
y = x**2
```

```python
# Creating the line plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = x^2', color='blue')
```

```python
# Customizing the line plot
plt.title('Line Plot of y = x^2', fontsize=14)
plt.xlabel('x values', fontsize=12)
plt.ylabel('y values', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
```

```python
# Displaying the line plot
plt.show()
```

```python
# Data for the bar plot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'green', 'blue', 'purple']
```

```python
# Creating the bar plot
plt.figure(figsize=(8, 6))
plt.bar(categories, values, color=colors)
```

```python
# Customizing the bar plot
plt.title('Bar Plot of Categories', fontsize=14)
plt.xlabel('Categories', fontsize=12)
plt.ylabel('Values', fontsize=12)
```

```python
# Displaying the bar plot
plt.show()
```

```python
# Data for the scatter plot
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
```

```python
# Creating the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x_scatter, y_scatter, color='orange')
```

```python
# Annotating each point
for i, (x_val, y_val) in enumerate(zip(x_scatter, y_scatter)):
    plt.text(x_val, y_val, f'({x_val}, {y_val})', fontsize=10, ha='right')
```

```python
# Customizing the scatter plot
plt.title('Scatter Plot with Annotations', fontsize=14)
plt.xlabel('x values', fontsize=12)
plt.ylabel('y values', fontsize=12)
```

```python
# Displaying the scatter plot
plt.show()
```

```python
# Data for the pie chart
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
```

```python
# Creating the pie chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['gold', 'skyblue', 'lightgreen', 'pink'])
```

```python
# Customizing the pie chart
plt.title('Pie Chart of Programming Languages', fontsize=14)
```

```python
# Displaying the pie chart
plt.show()
```
